### Bank Customer Churn Prediction

```python
# Churn Count
churn_count = data['churn'].value_counts()
# print(churn_count)

# Labels
labels = ['No Churn', 'Churn']
colors = ['green', 'red']


plt.figure(figsize=(6, 6))
plt.pie(
    churn_count,
    labels=labels,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    explode=(0, 0.1)
)
plt.title('Customer Churn Distribution')
plt.show()
```
