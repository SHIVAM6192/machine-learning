```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define function and its derivative 
def f(x):
    return x**2

def df(x):
    return 2*x

# Initial value
x = 5       # start point
alpha = 0.01 # learning rate
iterations = 2500

# Store values for visualization
x_history = [x]
y_history = [f(x)]

# Perform Gradient Descent
for i in range(iterations):
    grad = df(x)           # compute gradient (slope)
    x = x - alpha * grad   # update rule
    x_history.append(x)
    y_history.append(f(x))

# Plot function
x_vals = np.linspace(-6, 6, 200)
y_vals = f(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label="f(x) = x", color='blue')
plt.scatter(x_history, y_history, color='red', label="Gradient Descent steps")
plt.plot(x_history, y_history, '--', color='red')

plt.title("Gradient Descent Visualization")
plt.xlabel("x (parameter)")
plt.ylabel("f(x) (loss)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Final value of x: {x:.4f}")
print(f"Minimum value f(x): {f(x):.4f}")
```
