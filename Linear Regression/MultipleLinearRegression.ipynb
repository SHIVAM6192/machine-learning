```python
# Multiple Linear Regression Example
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python

data = {
    'Size': [2100, 1600, 2400, 1416, 3000, 1985, 1534, 1427, 1380, 2750],
    'Bedrooms': [3, 2, 4, 2, 4, 3, 3, 2, 3, 4],
    'Age': [20, 15, 30, 20, 8, 18, 24, 40, 30, 12],
    'Price': [500000, 340000, 620000, 320000, 700000, 460000, 380000, 290000, 310000, 650000]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Step 2: Split independent and dependent variables
X = df[['Size', 'Bedrooms', 'Age']]
y = df['Price']
```

```python
# Step 3: Create and train model
model = LinearRegression()
model.fit(X, y)
```

```python
# Step 4: Display coefficients
print("Intercept (b0):", model.intercept_)
print("Coefficients (b1, b2, b3):", model.coef_)
```

```python
# Step 5: Predict a new value
new_house = [[2500, 4, 10]]  # Size=2500 sqft, Bedrooms=4, Age=10 years
predicted_price = model.predict(new_house)
print("Predicted Price for 2500 sqft, 4 bed, 10 yrs old:", predicted_price[0])
```

```python
# Step 6 (optional): Visualize one feature’s effect
plt.scatter(df['Size'], df['Price'], color='blue')
plt.xlabel('Size (sqft)')
plt.ylabel('Price ($)')
plt.title('House Price vs Size')
plt.show()
```

```python
new_house = [[1000, 3, 5]]  # Size=2500 sqft, Bedrooms=4, Age=10 years
predicted_price = model.predict(new_house)
print("Predicted Price :", predicted_price[0])
```
